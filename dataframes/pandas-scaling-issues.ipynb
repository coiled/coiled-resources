```python
def read_taxi_csv(month):
    return pd.read_csv(
    f"/Users/powers/Documents/tmp/data/yellow_tripdata/yellow_tripdata_2019-{month}.csv",
    parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"],
    dtype={
        "payment_type": "UInt8",
        "VendorID": "UInt8",
        "passenger_count": "UInt8",
        "RatecodeID": "UInt8",
        "store_and_fwd_flag": "category",
        "PULocationID": "UInt16",
        "DOLocationID": "UInt16",
    },
)
```
